```python
import httpx
from bs4 import BeautifulSoup
import pandas as pd

# URL của trang web
url = "https://interviewkickstart-com.translate.goog/blogs/articles/machine-learning-mcqs-answers?_x_tr_sl=en&_x_tr_tl=vi&_x_tr_hl=vi&_x_tr_pto=tc"

# Tạo một SSLContext Cho phép legacy renegotiation (Vận dụng các lỗ hổng bảo mật của cấu hình SSL/TLS của máy chủ)
ssl_context = httpx.create_ssl_context()
ssl_context.options |= 0x4  # Cho phép legacy renegotiation

# Gửi response với SSLContext
with httpx.Client(verify=ssl_context) as client:
    response = client.get(url, headers={"User-Agent": "Mozilla/5.0"})
print(response.status_code)
# Kiểm tra response
if response.status_code == 200:
    content = response.text
else:
    print(f"Không đọc được trang đích: {response.status_code}")
    exit()

# ĐỌc content HTML bằng BeautifulSoup
soup = BeautifulSoup(content, 'html.parser')
```

```python
soup
```

```python
element = soup.find(attrs={"data-id":"31b2932"})
print(element)
```

```python
# Extract questions
questions = element.find_all('h3')

# Extract answers
answers = element.find_all('li')
dap_an = element.find_all('p')
dap_an = dap_an[4:]
# Initialize result list
result = []
```

```python
print(questions)
print(answers)
print(dap_an)
```

```python
# Lấy câu hỏi
question_ne = [question.text.strip() for question in questions]
answer_ne = [answer.text.strip() for answer in answers]
dap_an_ne = [dapan.text.strip() for dapan in dap_an]

print(question_ne)
print(answer_ne)
print(dap_an_ne)
```

```python
answer_ne = [answer_ne[i:i+4] for i in range(0, len(answer_ne), 4)]
print(answer_ne)
```

```python
print(question_ne)
print(answer_ne)
print(dap_an_ne)
```

```python
print(len(answer_ne))
print(len(question_ne))
print(len(dap_an_ne))
```

```python
dap_an_ne = [item for item in dap_an_ne if 'Answer' in item]
```

```python
print(len(answer_ne))
print(len(question_ne))
print(len(dap_an_ne))
```

```python
answer_ne = answer_ne[:-3]
question_ne = question_ne[:-3]
```

```python
print(len(answer_ne))
print(len(question_ne))
print(len(dap_an_ne))
```

```python
result = []
for i in range(len(question_ne)):
    question_data = {
        'Câu hỏi': question_ne[i],
        'Lựa chọn A': answer_ne[i][0],
        'Lựa chọn B': answer_ne[i][1],
        'Lựa chọn C': answer_ne[i][2],
        'Lựa chọn D': answer_ne[i][3],
        'Đáp án': dap_an_ne[i]
    }
    result.append(question_data)
print(result)
```

```python
import pandas as pd
df = pd.DataFrame(result)
# Ghi DataFrame vào file CSV
df.to_csv('cau_hoi_ml.csv', index=False, encoding='utf-8-sig')

print("Dữ liệu đã được ghi vào file 'cau_hoi_ml.csv' thành công!")
```
